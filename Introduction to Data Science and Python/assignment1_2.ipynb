```python
import pandas as pnd
import matplotlib.pyplot as plt
```

```python
happinessData=pnd.read_csv("happiness-cantril-ladder.csv")
alcoholData=pnd.read_csv("total-alcohol-consumption.csv")
suicideData=pnd.read_csv("annual-number-of-deaths-by-cause.csv")
```

```python
happinessData_usa = happinessData[(happinessData["Entity"] == "United States") & (happinessData["Year"] >= 2005)]
suicideData_usa = suicideData[(suicideData["Entity"] == "United States") & (suicideData["Year"] >= 2005)]
alcoholData_usa = alcoholData[(alcoholData["Entity"] == "United States") & (alcoholData["Year"] >= 2005)]
suicideData_usa_sort=suicideData_usa.sort_values(by='Year')
```

```python
fig, ax = plt.subplots(figsize=(14,7))
ax1 = ax.twinx()
ax.plot(happinessData_usa["Year"], happinessData_usa["Life satisfaction in Cantril Ladder (World Happiness Report 2022)"], color='b', label='Life satisfaction in Cantril Ladder (World Happiness Report 2022)')
ax1.plot(suicideData_usa_sort["Year"], suicideData_usa_sort["Deaths - Self-harm - Sex: Both - Age: All Ages (Number)"], color='r', label='Deaths - Self-harm - Sex: Both - Age: All Ages (Number)')
fs=15
plt.title("Happiness vs Suicide", fontsize = fs)
ax.set_xlabel("Year", fontsize = fs)
ax.set_ylabel("Life satisfaction in Cantril Ladder (World Happiness Report 2022)", fontsize = fs)
ax1.set_ylabel("Deaths - Self-harm - Sex: Both - Age: All Ages (Number)", fontsize = fs)
fig.legend(loc='lower left', bbox_to_anchor=(0.13,-0.2))
```

```python
fig, ax = plt.subplots(figsize=(14,7))
ax1 = ax.twinx()
ax.plot(alcoholData_usa["Year"], alcoholData_usa["Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)"], color='b', label='Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)')
ax1.plot(suicideData_usa_sort["Year"], suicideData_usa_sort["Deaths - Self-harm - Sex: Both - Age: All Ages (Number)"], color='r', label='Deaths - Self-harm - Sex: Both - Age: All Ages (Number)')
fs=15
plt.title("Alcohol consumption vs Suicide", fontsize = fs)
ax.set_xlabel("Year", fontsize = fs)
ax.set_ylabel("Total alcohol consumption (liters of pure alcohol per person)", fontsize = fs)
ax1.set_ylabel("Deaths - Self-harm - Sex: Both - Age: All Ages (Number)", fontsize = fs)
fig.legend(loc='lower left', bbox_to_anchor=(0.13,-0.2))
```
